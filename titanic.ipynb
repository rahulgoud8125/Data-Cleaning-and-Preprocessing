```python
def ad(x):
  if x== 'Amazfit':
    return 181
  elif x== 'Apple':
    return 523
  elif x=='Fitbit':
    return 204
  elif x=='Garmin':
    return 583
  elif x=='Huawei':
    return 465
  elif x=='Oura':
    return 427
  elif x=='Polar':
    return 342
  elif x=='Samsung':
    return 438
  elif x=='WHOOP':
    return 30
  else:
    return 354
```

```python
def da(x):
  if x== 'Amazfit':
    return 3.2
  elif x== 'Apple':
    return 3.3
  elif x=='Fitbit':
    return 2.9
  elif x=='Garmin':
    return 2.5
  elif x=='Huawei':
    return 3.2
  elif x=='Oura':
    return 2.5
  elif x=='Polar':
    return 3.
  elif x=='Samsung':
    return 3.2
  elif x=='WHOOP':
    return 2.1
  else:
    return 2.5
```

```python
x['GPS_Accuracy_Meters']=x['Brand'].apply(da)
```
